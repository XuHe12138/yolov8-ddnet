```python
video=cv2.VideoCapture("./ultralytics/assets/ddnet-train1.mp4")
num=0
save_step=10
while True:
    ret,frame=video.read()
    if not ret:
        break
    num+=1
    if num%save_step==0:
        cv2.imwrite("./demo_images/"+str(num)+".jpg",frame)
```
